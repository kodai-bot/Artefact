```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the data
tuatha_nw = gpd.read_file('/Users/mensab/Documents/GISMapFiles/Files1600/baronies_NW.shp')
```

```python
inishowen_east = tuatha_nw[tuatha_nw['ENGLISH'] == 'INISHOWEN EAST']
inishowen_east.plot()
```

```python
inishowen_west = tuatha_nw[tuatha_nw['ENGLISH'] == 'INISHOWEN WEST']
inishowen_west.plot()
```

```python
Inis_Eogan = gpd.GeoSeries(inishowen_east.geometry, inishowen_west.geometry)
Inis_Eogan
```

```python
Inis_Eogan_union = Inis_Eogan.unary_union
Inis_Eogan_union
```

```python
#create a geodataframe from the union
Inis_Eogan_union = gpd.GeoDataFrame(geometry=[Inis_Eogan_union])
Inis_Eogan_union.plot()
```
