```python
# Assign the rows in tuath_nw to a variable to be deleted

regions_to_replace = ['INISHOWEN EAST', 'INISHOWEN WEST', 'RAPHOE NORTH', 'RAPHOE SOUTH', 'CORRAN', 'COOLAVIN', 'LOUGHTEE UPPER', 'LOUGHTEE LOWER', 'BOYLAGH', 'BANAGH']
```

```python
# Delete the old entries
tuatha_nw = tuatha_nw[~tuatha_nw['ENGLISH'].isin(regions_to_replace)]
```

```python
# Append the new unified geometries
# For 'INISHOWEN EAST' and 'INISHOWEN WEST'

Inis_Eogan_union['ENGLISH'] = 'INISHOWEN'
tuatha_nw = pd.concat([tuatha_nw, Inis_Eogan_union])

Cinel_Enna_union['ENGLISH'] = 'RAPHOE'
tuatha_nw = pd.concat([tuatha_nw, Cinel_Enna_union])

# For 'CORRAN' and 'COOLAVIN'
Corann_union['ENGLISH'] = 'CORANN'
tuatha_nw = pd.concat([tuatha_nw, Corann_union])

# For 'LOUGHTEE UPPER' and 'LOUGHTEE LOWER'
Loughtee_union['ENGLISH'] = 'LOUGHTEE'
tuatha_nw = pd.concat([tuatha_nw, Loughtee_union])

# For 'BOYLAGH' and 'BANAGH'
Boylagh_union['ENGLISH'] = 'BOYLAGH'
tuatha_nw = pd.concat([tuatha_nw, Boylagh_union])

```

READ PARISHES, THESE ARE IN THEORY SUBDIVISIONS OF BARONIES, POSSIBLY REPRESENTING SUBCHIEFS.


```python
civil_parishes = gpd.read_file('/Users/mensab/Documents/GISMapFiles/civil_parishes')
civil_parishes.head()
```

```python
# CRS is EPSG 4326, which is WGS 84, AND NEEDS TO BE CHANGED TO EPSG 2157, WHICH IS IRISH GRID.
# civil_parishes.crs

civil_parishes = civil_parishes.to_crs(epsg=2157)

civil_parishes.crs
```

```python
# Reset the index
civil_parishes = civil_parishes.reset_index(drop=True)
```

```python
# Perform intersection between civil_parishes and tuatha_nw_1600
# This will give us the civil parishes that are within the tuatha_nw_1600 shapefile.

parishes_nw = gpd.overlay(tuatha_nw_1600, civil_parishes, how='intersection')

```

```python
parishes_nw.info()
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

tuatha_nw_1600.plot(ax=ax, color='white', edgecolor='black',linewidth = 3, zorder=1)
parishes_nw.plot(ax=ax, column = 'AREA_2', cmap = 'Greys_r', facecolor='Gray', alpha = 0.4, edgecolor='black', zorder=2)


plt.show()

```
